```python
# Importing necessary libraries
import numpy as np  # For numerical operations
import seaborn as sns  # For creating plots with enhanced aesthetics
import matplotlib.pyplot as plt  # For plotting

# Update the default matplotlib settings for font and math text
params = {'mathtext.default': 'regular', 'font.family': 'C059'}          
plt.rcParams.update(params)

# Generate data for the curves
# x is a range of values for the horizontal axis (Quantity)
x = np.linspace(0, 6, 400)  # Start from 0, 400 points in the range
y1 = x**2  # Concave up curve (representing total utility)
y2 = -x**2 + 27*x - 1.2*x**2  # Concave down curve (representing production cost)

# Additional data for price curve
# x_price0 is a range representing the price curve, from 3 to 7
x_price0 = np.linspace(3, 7, 400)
y_price0 = 74 + 2.25*x  # Linear price curve equation

# Create the plot with a specified figure size
plt.figure(figsize=(8, 6))

# Plotting the concave down production cost curve
sns.lineplot(x=x, y=y2, color='gray', ls='--')  # Dashed gray line for production cost

# Plotting the concave up utility curve
sns.lineplot(x=x, y=y1, color='black', ls='--')  # Dashed black line for utility

# Adding text annotations for curves
plt.text(6.1, 80, 'W (Utility)', fontsize=14)  # Label for utility curve
plt.text(6.1, 30, 'C (Production Cost)', fontsize=14)  # Label for production cost curve

# Plotting the price curve in red
sns.lineplot(x=x_price0, y=y_price0, color='red')  # Red line for price curve

# Adding text annotation for price curve and coordinates
plt.text(7.1, 85, r'$P_0$', fontsize=14, color='red')  # Label for price curve
plt.text(4.7, 75, r'$(Q_0, W_0)$', fontsize=14, color='black')  # Point label for (Q_0, W_0)

# Replotting the concave down production cost curve to ensure it's visible (appears twice)
sns.lineplot(x=x, y=y2, color='gray', ls='--')

# Plot a specific point on the graph (e.g., equilibrium or intersection point)
plt.plot(5.4, 82, 'ko')  # 'ko' means black circle marker at (5.4, 82)

# Customizing the plot
plt.title("", fontsize=16)  # No title
plt.xlabel("Quantity Consumer or Produced", fontsize=16)  # Label for x-axis
plt.xticks([])  # Removing x-axis ticks
plt.ylabel("Total Utility or Production Cost ($)", fontsize=16)  # Label for y-axis
plt.yticks([])  # Removing y-axis ticks

# Adding horizontal and vertical lines through the origin
plt.axhline(0, color='black', linewidth=1)  # Horizontal axis (x=0) line
plt.axvline(0, color='black', linewidth=1)  # Vertical axis (y=0) line

# Set the limits for the plot axes
plt.xlim(0, 10)  # x-axis limits from 0 to 10
plt.ylim(0, 100)  # y-axis limits from 0 to 100

# Remove the top and right spines (borders of the plot)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Show the plot
plt.show()
```
